### Use of isnull()

```python
dataset.isnull().sum()
```

## Keep only columns where 80% or more valid data is available

```python
df=dataset.dropna(axis=1,thresh=dataset.shape[0]*0.8)
```

```python
df.shape
```

```python
df.head(8)
```

## Filling null values

Filling null values with a previous values

```python
df1 = df.fillna(method = 'pad')
```

```python
df1.head(8)
```

Filling null values with next value

```python
df2 = df.fillna(method = 'bfill')
df2.head(9)
```

Filling null values with median

```python
df3= df.fillna(value=df['LotFrontage'].median())
df3.head(8)
```

# Searching and Filtering with Pandas

Filter only those data whose Sale Price >= 500000


```python
filt = (df3['SalePrice']>= 500000)
```

```python
df3.loc[filt, 'SalePrice']
```

Filter only those data whose Street is Pave and LotShape is IR3

```python
filt2 = (df3['Street']=='Pave') & (df3['LotShape'] == 'IR3')
```

```python
df3.loc[filt2, 'YrSold']
```

Filter only those data whose Sale Condition is Abnormal or Sale Price > 500000

```python
filt3 = (df3['SaleCondition']=='Abnormal') | (df3['SalePrice'] > 500000)
```

```python
df3.loc[filt3, ['YrSold','SaleCondition','SalePrice']]
```
